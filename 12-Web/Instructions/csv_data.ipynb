```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
file_to_load = "Resources/cities.csv"
```

```python
data_csv = pd.read_csv(file_to_load)
data_csv.head()
                  
```

```python
data_df = data_csv[["City_ID",'City','Cloudiness','Country','Date', 'Humidity','Lat','Lng','Max Temp','Wind Speed']]
data_df
```

```python
plt.figure(figsize=(10,8))
ax=sns.scatterplot(x='Lat',y='Max Temp',hue="Max Temp", data=data_df,)

plt.suptitle("Max Temp VS Latitude", size=20)
plt.legend(loc='upper left', fontsize=15, bbox_to_anchor=(1, 1.2))
plt.xlabel("Latitude", size= 15)
plt.ylabel("Max Temperature", size=15)
plt.xlim(-60,90)
plt.ylim(0,120)
plt.show()

```

```python
plt.plot()
ax=sns.scatterplot(x='Lat',y='Cloudiness',hue="Max Temp", data=data_df,hue_norm=(20, 110))
plt.suptitle("Max Temp VS Latitude", size=20)
plt.legend(loc='upper left', fontsize=15, bbox_to_anchor=(1, 1.2))
plt.xlabel("Latitude", size= 15)
plt.ylabel("Max Temperature", size=15)
plt.show()
```
